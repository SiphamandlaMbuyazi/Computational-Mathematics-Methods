This is the function that is responsible for `rotations` if you know the `Jacobi` algorithm then you know it should have rotation matrix say `R` in every say `i-th` iteration, 

this `jacobi_rotation` function will then calculates the `rotation matrix  R` for the largest off-diagonal element

```python
def jacobi_rotation(A, i, j):
    if A[i, i] == A[j, j]:
        theta = np.pi / 4
    else:
        theta = 0.5 * np.arctan(2 * A[i, j] / (A[i, i] - A[j, j]))
    
    c = np.cos(theta)
    s = np.sin(theta)
    #this is just building the rotation matrix as we know it right, this is useful when doing the similar transformations
    R = np.identity(len(A))
    R[i, i] = c
    R[j, j] = c
    R[i, j] = s
    R[j, i] = -1*s
    
    return R
```

Now we're ready to define the `Jacobi method` algorithm, this will be responsible to then perform the iterations, see the function below.

this function takes in the matrix, and there's also the `maximum iterations`, and the `tolerance` which is free to be changed and set to anything

```python
def jacobi_method(A, tol=1e-200000, max_iterations=430):
    n = len(A)
    eigenvectors = np.identity(n)
    
    for it in range(max_iterations):
        # Find the largest off-diagonal element
        off_diagonal = np.tril(A, -1)
        i, j = np.unravel_index(np.argmax(np.abs(off_diagonal)), A.shape)
        
        # Stop if the off-diagonal elements are below the tolerance
        #stop when the off-diagonal elements are zero.
        if np.abs(A[i, j]) ==0:
            break
        
        # Compute the rotation matrix
        R = jacobi_rotation(A, i, j)
        
        # Update the matrix A
        #A = R.T @ A @ R #@ is just for matrix maltiplications, I was recently released, so basically to multiply the matrices say A and B, in numpy you do np.dot(A,B) the short hand for this which is more efficient, this is just shorthand and should not be intimidating
        print(f"{it} : \n {A}")

        B=np.dot(R.T,A)
        A=np.dot(B,R)
        # Update the eigenvectors
        eigenvectors = eigenvectors @ R
    
    eigenvalues = np.diag(A)
    return eigenvalues, eigenvectors
```
